## üßπ Step 3: Data Cleaning



```python
df.dropna(subset=['rating', 'duration', 'date_added'], inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', infer_datetime_format=True)

df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df.isnull().sum()
```

## üìà Step 4: Exploratory Data Analysis
### - Type Distribution
### - Top Countries
### - Genre Breakdown

```python
#  Movie vs TV Show count
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Content Type: Movies vs TV Shows")
plt.xlabel("")
plt.ylabel("Count")
plt.show()

```

```python
# Top 10 countries producing Netflix content
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', figsize=(8,5), color='orange')
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
#  Content released over the years
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5), color='teal')
plt.title("Netflix Releases by Year")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.grid(True)
plt.show()
```

```python
#  Content added to Netflix each year
df['year_added'].value_counts().sort_index().plot(kind='bar', figsize=(8,4), color='plum')
plt.title("Content Added to Netflix Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
#  Distribution of Content Ratings
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index, palette='coolwarm')
plt.title("Distribution of Content Ratings on Netflix")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

```python
# Genre (Category) Analysis

from collections import Counter

# Split and count the genres
genre_data = df['listed_in'].dropna().apply(lambda x: x.split(', '))
genre_flat = [genre for sublist in genre_data for genre in sublist]
top_genres = Counter(genre_flat).most_common(10)

# Convert to DataFrame for plotting
genre_df = pd.DataFrame(top_genres, columns=['Genre', 'Count'])

# Plot
sns.barplot(data=genre_df, y='Genre', x='Count', palette='mako')
plt.title("Top 10 Netflix Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
# Content Duration Analysis

# Separate movies and shows
movies = df[df['type'] == 'Movie']
shows = df[df['type'] == 'TV Show']

# Clean movie durations (remove ' min')
movies['duration'] = movies['duration'].str.replace(' min', '').astype(float)

# Count of most common durations
movies['duration'].value_counts().head(10).plot(kind='bar', color='coral')
plt.title("Top Movie Durations")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

```python
# Most Frequent Directors
top_directors = df['director'].dropna().value_counts().head(10)
top_directors.plot(kind='barh', color='darkgreen')
plt.title("Top 10 Most Frequent Directors")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.show()
```

```python
# Most Frequent Actors
cast_data = df['cast'].dropna().apply(lambda x: x.split(', '))
cast_flat = [actor.strip() for sublist in cast_data for actor in sublist]
top_cast = Counter(cast_flat).most_common(10)

cast_df = pd.DataFrame(top_cast, columns=['Actor', 'Count'])

sns.barplot(data=cast_df, y='Actor', x='Count', palette='rocket')
plt.title("Top 10 Most Featured Actors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Actor")
plt.show()
```
